```python
import pandas as pd
import numpy as np
import requests
import datetime
import json
```

```python
url="https://api.spacexdata.com/v4/launches/past"
```

```python
def request_data(url): 
    req = requests.get(url)
    req_jason = req.json() # convert the json result into a dataframe
    return req_jason
```

```python
req_jason = request_data(url)
```

```python
data = pd.json_normalize(req_jason)
data.columns
```

```python
data = data[['rocket','payloads','launchpad', 'flight_number', 'date_utc', 'cores']]
data
```

```python
data['payloads']
```

```python
# Delete rows with multiple payloads in one rocket
data = data[data['payloads'].map(len)==1]
```

```python
data['payloads']
```

```python
data['payloads'] = data['payloads'].map(lambda x : x[0])
```

```python
data['payloads']
```

```python
data['cores'][0]
```

```python
data['cores']
```

```python
# Delete rows with multiple cores.
data = data[data['cores'].map(len)==1]
```

```python
data['cores'] = data['cores'].map(lambda x : x[0])
```

```python
data['cores']
```

```python
data['date'] = pd.to_datetime(data['date_utc']).dt.date
data['date']
```

```python
# Using the date we will restrict the dates of the launches
data = data[data['date'] <= datetime.date(2023, 4, 30)]
```

```python
data
```

```python
response_1 = requests.get("https://api.spacexdata.com/v4/rockets/").json()
```

```python
response_1[0]
```

```python
BoosterVersion = []
for i in data['rocket']:
    response_1 = requests.get("https://api.spacexdata.com/v4/rockets/"+str(i)).json()
    BoosterVersion.append(response_1['name'])
BoosterVersion
```

```python
response_2 = requests.get("https://api.spacexdata.com/v4/payloads/").json()
```

```python
response_2[0]
```

```python
PayloadMass=[]
Orbit=[]
for j in data['payloads']:
    response_2 = requests.get("https://api.spacexdata.com/v4/payloads/"+ str(j)).json()
    PayloadMass.append(response_2['mass_kg'])
    Orbit.append(response_2['orbit']) 
PayloadMass,Orbit
```

```python
response_3 = requests.get("https://api.spacexdata.com/v4/launchpads/").json()
```

```python
response_3[0]
```

```python
LaunchSite = []
Latitude = []
Longitude = []

for x in data['launchpad']:
    response_3 = requests.get("https://api.spacexdata.com/v4/launchpads/"+str(x)).json()
    LaunchSite.append(response_3['name'])
    Latitude.append(response_3['latitude'])
    Longitude.append(response_3['longitude'])
    
LaunchSite,Latitude,Longitude
```

```python
response_4 = requests.get("https://api.spacexdata.com/v4/cores/").json()
```

```python
response_4[0]
```

```python
Block = []
ReusedCount = []
Serial = []

for y in data['cores']:
    response_4 = requests.get("https://api.spacexdata.com/v4/cores/"+ y['core']).json()
    Block.append(response_4['block'])
    ReusedCount.append(response_4['reuse_count'])
    Serial.append(response_4['serial'])
    
Block,ReusedCount,Serial
```

```python
data['cores'][0]
```

```python
Flights = []
GridFins = []
Legs = []
Reused = []
Outcome = []
LandingPad = []

for z in data['cores']:
    Flights.append(z['flight'])
    GridFins.append(z['gridfins'])
    Legs.append(z['legs'])
    Reused.append(z['reused'])
    Outcome.append(str(z['landing_success'])+' '+str(z['landing_type']))
    LandingPad.append(z['landpad'])

Flights, GridFins, Legs, Reused, Outcome, LandingPad
```

```python
space_X_rawdata={}

space_X_rawdata['FlightNumber'] = data['flight_number']
space_X_rawdata['Date'] = data['date']
space_X_rawdata['BoosterVersion'] = BoosterVersion
space_X_rawdata['PayloadMass'] = PayloadMass
space_X_rawdata['Orbit'] = Orbit
space_X_rawdata['LaunchSite'] = LaunchSite 
space_X_rawdata['Outcome'] = Outcome
space_X_rawdata['Flights'] = Flights
space_X_rawdata['GridFins'] = GridFins
space_X_rawdata['Reused'] = Reused
space_X_rawdata['Legs'] = Legs
space_X_rawdata['LandingPad'] = LandingPad
space_X_rawdata['Block'] = Block
space_X_rawdata['ReusedCount'] = ReusedCount
space_X_rawdata['Serial'] = Serial
space_X_rawdata['Longitude'] = Longitude
space_X_rawdata['Latitude'] = Latitude
```

```python
space_X_rawdata
```

```python
df = pd.DataFrame(space_X_rawdata)
df
```

```python
df.to_csv('Space_X.csv', index=False)
```
